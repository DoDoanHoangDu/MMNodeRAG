```python
all_synonyms = []
response_entities = ['prevention']
for ent in response_entities:
    ent = ent.upper()
    if ent not in graph_entities:
        continue
    print(ent)
    entity_node_id = graph_entities[ent]
    print(entity_node_id)
    entity_node = g6[entity_node_id]
    print(entity_node)
    print(entity_node.source)
    if not entity_node.source:
        continue
    synonym_ids = list(entity_node.source.split(","))
    print("Synonym IDs:",synonym_ids)
    synonyms = [g6[synonym_id].content for synonym_id in synonym_ids]
    all_synonyms.extend(synonyms)
for syn in all_synonyms:
    print(syn)
```
